## **Definição TensorFlow:**

 **TensorFlow é uma biblioteca de código aberto criada para Machine Learning, Deep Learning e Computação Númerica.** 

Também conhecido como framework, é uma multi-plataforma e pode ser usado no Windows, Linux, Mac OS, bem como executado em GPU's, CPU's, ou ainda em TPU's. Com API para diversas linguagem de programação.

Foi desenvolvido pelo Google em 2015, e o estilo do TensorFlow é bem semelhante a forma de programação da biblioteca Numpy do python.

## **GPU:**

**CPU X GPU:**

Temos o TensorFlow para as duas opções, por isso é importante entender a diferença entre elas.

**CPU:** é a unidade de processamento central. É um microprocessador de entrada e saída de dados.(Em linhas gerais é o cérebro dos computadores).

**GPU:** é a unidade de processamento gráfico. São processadores para programação paralela.

8.  **TensorFlow Research Cloud:** Programa de cluster para pesquisas.

**Como funciona:**

Funciona basicamente da seguinte maneira:

**1º** Precisamos criar um *blueprint* do código, tipo um esqueleto do que será executado (como uma variável,ou instância) por exemplo.

**2º** Depois cria uma sessão para executar o código, com o ```tf.Session()```. 
Com isso criamos um **grafo computacional** e enquando a *Session* não for aberta o código não irá fluir.

TensorFlow possui funções e bibliotecas já prontas para criação de todos os tipos de rede neurais, formando um ecossistema bem robusto para trabalhar com Deep Learning. Por isso, é simples e fácil a criação de uma rede neural com esse framework. Basta apenas realizar a importação dos módulos corretos e as configurações adequadas.

**TensorFlow**, significa *Tensor* - Que é o objeto matemático que representa o conjunto de dados. e *Flow* - O fluxo de operações que serão realizada nos tensores.

Esse fluxo é os **Grafos Computacionais**.

## **Grafos:**

maxresdefault.jpg

**Grafo Computacional:**

**Grafo** são estruturas matemáticas que permitem codificar relacionamentos entre pares de objetos.

**Os objetos são:** "nós"(Vertice) do grafo.

**Relacionamentos:** São as arestas.

Um Grafo, então é um modelo matemático que representa relações entre objetos. Por exemplo, um grafo G = (V, E) consiste de um conjunto de vértices V (também chamados de nós), ligados por um conjunto de arestas E.

file.png

Os grafos pode ser:


*   **Dirigidos / Direcionados:** As relações tem sentido definido. E as arestas são pares ordenados de vertices. Mesmo que sejam *self-loop* (Saindo e chegando nele mesmo.)
*   **Não dirigidos / Não Direcionado:** As relações não tem sentido definido. As arestas são pares não ordenados e não permite  *self-loop*

Para criar um grafo computacional, fazemos cada uma dessas operações nos nós, juntamente com as variáveis de entrada. Quando o valor de um nó é a entrada para outro nó, uma seta vai de um para outro e temos nesse caso um grafo direcionado.

Esses tipos de grafos surgem o tempo todo em Ciência da Computação, especialmente ao falar sobre programas funcionais. Eles estão intimamente relacionados com as noções de grafos de dependência e grafos de chamadas. Eles também são a principal abstração por trás do popular framework de Deep Learning, o **TensorFlow.**





## **Tensores:**

tensor.png

**TENSOR:**



Tensor é um objeto matemático. É basicamente um vetor N-dimensional, significando que um tensor pode ser usado para representar conjuntos de dados N-dimensionais. Como na imagem acima.

À medida que a dimensão continua crescendo, os dados se tornam cada vez mais complexos. Por exemplo, se pegarmos um Tensor da forma (3x3), posso chamá-lo de matriz de 3 linhas e 3 colunas. Os tensores podem ser constantes ou variáveis.

Por exemplo:

Um conjunto de dados com **sem dimensão**, chamamos de **Escalar;** No TensorFlow é o **Rank 0.**

Um conjunto de dados **com 1 dimensão**, chamamos de **Vetor;** No TensorFlow é o **Rank 1.**

Um conjunto de dados **com 2 dimensões**, chamamos de **Matriz;** No TensorFlow é o **Rank 2.**

Um conjunto de dados **com + de 3 dimensões**, chamamos de **Tensor.** Nesse caso é o **Rank 3.**


**TensorFlow x Numpy**

TensorFlow tem muitas semelhanças com o numpy, dentre as principais diferenças estão as limitações do Numpy para criar funções de tensor e automaticamente computar derivadas além de não ter suporte a GPU. Mas, apesar dessas limitações é um excelente e indispensável pacote para análise de dados e computação científica.

## **Modelo de Programação TensorFlow**


O principal objetivo de um programa TensorFlow é expressar uma computação numérica como um grafo direcionado atráves da fórmula de **h = ReLU (Wx + b)**.

Este é um componente é clássico em muitas redes neurais, que conduz a transformação linear dos dados de entrada e, em seguida, alimenta uma linearidade (função de ativação linear retificada, neste caso).


Podemos ver o grafo como um sistema, que tem entradas (**os dados x**), saída (**h neste caso**), variáveis com estado (**W e b**) e um monte de operações (**matmul, add e ReLU**).

**Variáveis:** quando treinamos um modelo, usamos variáveis para manter e atualizar parâmetros. Ao contrário de muitos tensores que fluem ao longo das margens do grafo, uma variável é um tipo especial de operação. Na maioria dos modelos de aprendizado de máquina, existem muitos parâmetros que temos que aprender, que são atualizados durante o treinamento. Variáveis são nós com estado que armazenam parâmetros e produzem seus valores atuais de tempos em tempos. Seus estados são mantidos em múltiplas execuções de um grafo. Por exemplo, os valores desses nós não serão atualizados até que uma etapa completa de treinamento usando um mini lote de dados seja concluída.

**Operações matemáticas:** Neste grafo, existem três tipos de operações matemáticas: 

A operação **MatMul** multiplica dois valores de matriz, na fórmula o 'W e X'; 

A operação **Add** adiciona elementos a operação, na fórmula o 'B';

**ReLU** é ativada com a função linear retificada de elementos.

### **Operações Matemáticas:**

**Soma:**

````python
#Criando o tensor como uma constante

const_a = tf.constant(5)
const_b = tf.constant(9)
````

````python
# Visualizando a variável constante

const_a
````

````python
# soma

total = const_a + const_b
print(total)
````

````python
# Criando nodes no Grafo Computacional

node1 = tf.constant(5, dtype = tf.int32) 
node2 = tf.constant(9, dtype = tf.int32)
#Aqui já definimos o tipo de dados que queremos com o 'dtype'
node3 = tf.add(node1, node2) #Forma mais comum de adição

#Visualizando a saída do grafo:

print("\nA soma do node1 e node2 é:",node3)
````

**Subtração:**

````python
#Criando um Tensor Randômico (aleatório)

rand_a = tf.random.normal([3], 2)
rand_b = tf.random.uniform([3], 1, 4)
````

````python
print(rand_a)
print(rand_b)
````

````python
# Função de subtração:

diff = tf.subtract(rand_a, rand_b)
````

````python
print(diff)
````

**Divisão:**

````python
node3 = tf.constant(21)
node4 = tf.constant(7)
````

````python
division = tf.math.truediv(node3,node4)
````

````python
print(division)
````

**Multiplicação:**

````python
# Criando o tensor:

tensor_c = tf.constant([[4., 2.]])
tensor_d = tf.constant([[3.], [7.]])
````

````python
tensor_c
````

````python
tensor_d
````

````python
# Multiplicação
# tf.math.multiply(X, Y) executa multiplicação element-wise 
# https://www.tensorflow.org/api_docs/python/tf/math/multiply

prod = tf.math.multiply(tensor_c, tensor_d)
````

````python
print('\nProduto Element-wise Entre os Tensores C e D: \n', prod)
````

````python
# Outra operação de Multiplicação de Matrize

mat_a = tf.constant([[2, 3], [9, 2], [4, 5]])
mat_b = tf.constant([[6, 4, 5], [3, 7, 2]])
````

````python
mat_a
````

````python
mat_b
````

````python
# Multiplicação
# tf.linalg.matmul(X, Y) executa multiplicação entre matrizes 
# https://www.tensorflow.org/api_docs/python/tf/linalg/matmul

mat_prod = tf.linalg.matmul(mat_a, mat_b) #A função 'linalg' é vem de Álgebra Linear
````

````python
print('\nProduto Element-wise Entre os Tensores (Matrizes): \n', mat_prod)
````

### **Criando um Tensor Variável:**

````python
# Criando o Tensor:

node5 = tf.Variable([10,7]) #Observar o módulo com letra maiúscula.
print(node5)
````

Quando criamos um Tensor como uma constante (```tf.constant```) ele se torna imutável, logo não pode sofrer nenhuma alteração nem novas atribuições de valores.

Porém, ao criarmos um Tensor como variável (```tf.Variable```) é possível modificá-los e/ou atribuir novos valores utilizando a função: ``` assign``` 

````python
node5[0].assign(7)
node5
````

Observe que o valor no índice ```0``` era **10**, e agora é **7**.

````python
print("Tipo de dado de cada elemento:", node8.dtype)
print("Número de dimensões (rank):", node8.ndim)
print("Shape do tensor:", node8.shape)
print("Elementos no eixo 0 do tensor:",node8.shape[0])
print("Elementos no último eixo do tensor", node8.shape[-1])
print("Número total de elementos (2*3*4*5):",tf.size(node8).numpy())
````
